```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as p1
from sklearn import linear_model
```

```python
data=pd.read_csv("/Users/aravinth/Desktop/Data Warehousing/Linear Regression/Gold_Price.csv")
```

```python
x=data.Year
y=data.Price
p1.title('Gold Price - Linear Regression')
p1.xlabel('Years')
p1.ylabel('Price')
p1.scatter(x,y,color='g',linestyle='solid')
p1.show()
```

```python
l=linear_model.LinearRegression()
l.fit(data[['Year']],data.Price)
l.predict([[2021]])
```

```python
print(l.coef_)
```

```python
print(l.intercept_)
```

```python
slope=l.coef_
intercept=l.intercept_
s=data.Year
d=data.Price
p1.scatter(s,d)
p1.plot(s,slope*s+intercept,color='g')
p1.show()

```
