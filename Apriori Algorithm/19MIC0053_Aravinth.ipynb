<h1>Apriori Algorithm

```python
import numpy as np
import pandas as pd
from apyori import apriori
vegetable=pd.read_csv("/Users/aravinth/Desktop/Data Warehousing/Assignment 1/Vegetable.csv")
print(vegetable)
```

```python
list1=[]
for i in range(0,30):
    list1.append([str(vegetable.values[i,j]) for j in range (0,10)])
    print(list1)
```

<h1>Association Rule

```python
association_rule=apriori(list1,min_support=0.5,min_confidence=1,min_length=2)
result= list(association_rule)
print(len(result))
print(result)
```

<h1>FP Growth Algorithm

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
automobile=pd.read_csv("/Users/aravinth/Desktop/Data Warehousing/Assignment 1/Automobile.csv")
list1=[]
for i in range (0,11):
    list1.append([str(automobile.values[i,j]) for j in range (0,7)])
print(list1)
trans=TransactionEncoder()
trans_ary=trans.fit(list1).transform(list1)
data=pd.DataFrame(trans_ary,columns=trans.columns_)
data
```

```python
from mlxtend.frequent_patterns import fpgrowth
fpgrowth(data,min_support=0.2)
```

```python
frequent_item=fpgrowth(data,min_support=0.2,use_colnames=True)
print(frequent_item)
```

<h1>Association Rule

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_item,metric='confidence',min_threshold=0.5)
rules
```
